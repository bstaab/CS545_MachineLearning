# Multilabel Classification

We will maximize the log likelihood of the training data.
$$
  \begin{align*}
    L(\Wv) & = \left ( \prod_{n=1}^N \prod_{k=1}^K P(C=k|\xv_n)^{t_{n,k}} \right ) \\
    \log L(\Wv) & = LL(\Wv) =  \sum_{n=1}^N \sum_{k=1}^K t_{n,k} \log P(C=k|\xv_n)\\
     & =  \sum_{n=1}^N \sum_{k=1}^K t_{n,k} \log g_{n,k}\\
     g_{n,k}(s_{n,k}) &= \frac{1}{1 + e^{-s_{n,k}}}, \text{ where } s_{n,k} \text{ is weight sum of inputs}
   \end{align*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(40).reshape((-1, 1))
T = np.hstack((X % 2 == 0, X % 3 == 0)).astype(int)
X.shape, T.shape
```

```python
plt.figure(figsize=(15, 2))
plt.plot(T[:, 0], 's', ms=10, alpha=0.8)
plt.plot(T[:, 1], 'D', ms=10, alpha=0.8)
plt.legend(['div 2', 'div 3'])
```
